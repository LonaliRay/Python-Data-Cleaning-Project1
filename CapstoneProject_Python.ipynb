```python
project = pd.read_csv("project.csv")
employee = pd.read_csv("employee.csv")
seniority = pd.read_csv("seniority.csv")
```

```python
#2 fill missing cost using average(with loop)
avg_cost = project["Cost"].mean()

for i in range(len(project)):
    if pd.isna(project.loc[i, "Cost"]):
        project.loc[i, "Cost"] = avg_cost
print(avg_cost)
print(project)
```

```python
#3 split nameinto first name and last name
employee[["First Name", "Last Name"]] = employee["Name"].str.split(" ", expand=True)

employee.drop("Name", axis=1, inplace=True)
print(employee)
```

```python
#4 Merge all three dataframes
final = pd.merge(project, employee, on="ID")
final = pd.merge(final, seniority, on="ID")
print(final)
```

```python
#5 Added a new bonus column in the Final dataframe. Give a 5% bonus concerning project 
#cost only to employees who have finished the projects.
final["Bonus"] = np.where(final["Status"] == "Finished", final["Cost"] * 0.05, 0)
```

```python
#6Denote level if project failed
final.loc[final["Status"] == "Failed", "Designation Level"] -= 1

# remove if level > 4
final = final[final["Designation Level"] <= 4]
```

```python
#7Added Mr and Mrs
def add_title(row):
    if row["Gender"] == "M":
        return "Mr. " + row["First Name"]
    else:
        return "Mrs. " + row["First Name"]

final["First Name"] = final.apply(add_title, axis=1)

final.drop("Gender", axis=1, inplace=True)
```

```python
#8promote if age > 29
    final.loc[final["Age"] > 29, "Designation Level"] += 1
```

```python
#9 total project cost per eemployees
total_cost = final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

total_cost.rename(columns={"Cost": "Total Cost"}, inplace=True)
```
